```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import os
from sklearn.neighbors import NearestNeighbors
from scipy.spatial import ConvexHull


# ==================== CONFIGURATION ====================

def get_color_palette():
    """Return the discrete rainbow color palette."""
    return [
        "#882E72", "#D6C1DE", "#437DBF", "#92C5DE",
        "#F7F056", "#F4A736", "#DC050C", "#A5170E"
    ]


# ==================== GLOBAL COLORBAR SETUP ====================

def calculate_global_colorbar_params(wide_df, col_x, n_color_bins):
    """
    Calculate global colorbar parameters for consistent coloring across all subjects.
    NOW USES 'i' VALUES INSTEAD OF RADIUS FOR COLORING.
    
    Parameters:
    -----------
    wide_df : DataFrame
        Full dataset with all subjects
    col_x : str
        Column name for radius values (not used for color anymore)
    n_color_bins : int
        Number of color bins
        
    Returns:
    --------
    tuple : (global_i_min, global_i_max, actual_n_bins_global, 
             global_color_bins, global_tick_labels)
    """
    print("\nCalculating global 'i' value range for universal colorbar...")
    
    # Use min and max 'i' values across all subjects
    global_i_min = wide_df['i'].min()
    global_i_max = wide_df['i'].max()
    print(f"  Global 'i' range: {global_i_min:.4f} to {global_i_max:.4f}")
    
    # Pre-calculate global color bins based on 'i' values
    actual_n_bins_global = n_color_bins
    global_color_bins = np.linspace(
        global_i_min,
        global_i_max,
        actual_n_bins_global + 1
    )
    global_tick_labels = [
        f"{global_color_bins[i]:.2f}-{global_color_bins[i+1]:.2f}"
        for i in range(actual_n_bins_global)
    ]
    
    return (global_i_min, global_i_max, actual_n_bins_global, 
            global_color_bins, global_tick_labels)


def calculate_local_colorbar_params(df_plot, col_x, n_color_bins):
    """
    Calculate local colorbar parameters for a single subject.
    NOW USES 'i' VALUES INSTEAD OF RADIUS FOR COLORING.
    
    Parameters:
    -----------
    df_plot : DataFrame
        Data for a single subject/laterality
    col_x : str
        Column name for radius values (not used for color anymore)
    n_color_bins : int
        Number of color bins
        
    Returns:
    --------
    tuple : (local_i_min, local_i_max, actual_n_bins_local, 
             local_color_bins, local_tick_labels)
    """
    # Use min and max 'i' values for this subject
    local_i_min = df_plot['i'].min()
    local_i_max = df_plot['i'].max()
    
    # Pre-calculate local color bins based on 'i' values
    actual_n_bins_local = n_color_bins
    local_color_bins = np.linspace(
        local_i_min,
        local_i_max,
        actual_n_bins_local + 1
    )
    local_tick_labels = [
        f"{local_color_bins[i]:.2f}-{local_color_bins[i+1]:.2f}"
        for i in range(actual_n_bins_local)
    ]
    
    return (local_i_min, local_i_max, actual_n_bins_local, 
            local_color_bins, local_tick_labels)


# ==================== BINNING AND SELECTION ====================

def create_2d_bins(df_plot, col_x, col_persistence, n_bins):
    """
    Create bins for 2D persistence plot.
    
    Parameters:
    -----------
    df_plot : DataFrame
        Data to bin
    col_x : str
        X-axis column name
    col_persistence : str
        Y-axis column name
    n_bins : int
        Number of bins
        
    Returns:
    --------
    tuple : (x_bins, y_bins)
    """
    x_min, x_max = df_plot[col_x].min(), df_plot[col_x].max()
    y_min, y_max = df_plot[col_persistence].min(), df_plot[col_persistence].max()

    x_bins = np.linspace(x_min, x_max, n_bins + 1)
    y_bins = np.linspace(y_min, y_max, n_bins + 1)
    
    return x_bins, y_bins


def circle_overlaps_rectangle(x_center, y_center, radius, rect_x_min, rect_x_max, rect_y_min, rect_y_max):
    """Check if a circle overlaps with a rectangle."""
    # Find the closest point on the rectangle to the circle center
    closest_x = np.clip(x_center, rect_x_min, rect_x_max)
    closest_y = np.clip(y_center, rect_y_min, rect_y_max)
    
    # Calculate distance from circle center to closest point
    distance = np.sqrt((x_center - closest_x)**2 + (y_center - closest_y)**2)
    
    # Circle overlaps if distance is less than or equal to radius
    return distance <= radius


def select_starred_points(df_plot, col_x, col_persistence, x_bins, y_bins, 
                         x_bin_index, y_bin_index, radius_size):
    """
    Select points whose center is in the bin or whose radius overlaps the bin.
    
    Parameters:
    -----------
    df_plot : DataFrame
        Data with bin assignments
    col_x, col_persistence : str
        Column names for x and y axes
    x_bins, y_bins : array
        Bin edges
    x_bin_index, y_bin_index : int
        Selected bin indices
    radius_size : float
        Radius for overlap detection
        
    Returns:
    --------
    tuple : (center_in_bin, radius_overlap_only, starred_points)
    """
    # Get bin boundaries
    bin_x_min = x_bins[x_bin_index]
    bin_x_max = x_bins[x_bin_index + 1]
    bin_y_min = y_bins[y_bin_index]
    bin_y_max = y_bins[y_bin_index + 1]
    
    # Apply overlap check to all points
    df_plot['radius_overlaps_bin'] = df_plot.apply(
        lambda row: circle_overlaps_rectangle(
            row[col_x], row[col_persistence], radius_size,
            bin_x_min, bin_x_max, bin_y_min, bin_y_max
        ),
        axis=1
    )
    
    # Get points with center in bin
    center_in_bin = df_plot[
        (df_plot['x_bin'] == x_bin_index) & (df_plot['y_bin'] == y_bin_index)
    ].copy()
    
    # Get points with radius overlap only (not center)
    radius_overlap_only = df_plot[
        df_plot['radius_overlaps_bin'] & 
        ~((df_plot['x_bin'] == x_bin_index) & (df_plot['y_bin'] == y_bin_index))
    ].copy()
    
    # Get all starred points (center in bin OR radius overlaps)
    starred_points = df_plot[
        ((df_plot['x_bin'] == x_bin_index) & (df_plot['y_bin'] == y_bin_index)) |
        df_plot['radius_overlaps_bin']
    ].copy()
    
    print(f"\n2D Selection Summary:")
    print(f"  Points with center in bin: {len(center_in_bin)}")
    print(f"  Additional points with radius overlap: {len(radius_overlap_only)}")
    print(f"  Total starred points (for 3D clustering): {len(starred_points)}")
    
    return center_in_bin, radius_overlap_only, starred_points


# ==================== CLUSTERING ====================

def compute_nearest_neighbor_tiers(df_plot, starred_points, minPts_x, minPts_y):
    """
    Compute three-tier nearest neighbor clustering.
    
    Parameters:
    -----------
    df_plot : DataFrame
        All points for this subject/laterality
    starred_points : DataFrame
        Points selected from 2D plot
    minPts_x, minPts_y : int
        Number of neighbors for tier 2 and tier 3
        
    Returns:
    --------
    tuple : (tier1_indices, tier2_indices, tier3_indices, nbrs_x, df_plot_index_list)
    """
    print(f"\n3D Clustering (using starred points as seed):")
    
    # Get all points from this subject/laterality for neighbor search
    all_coords = df_plot[['x', 'y', 'z', 'i']].values
    starred_coords = starred_points[['x', 'y', 'z', 'i']].values
    
    # Create mapping from array position to DataFrame index
    df_plot_index_list = df_plot.index.tolist()
    
    # Tier 1: Starred points (center in bin OR radius overlap)
    tier1_indices = set(starred_points.index)
    
    # Tier 2: Find minPts_x nearest neighbors for each starred point
    nbrs_x = NearestNeighbors(n_neighbors=minPts_x, algorithm='auto').fit(all_coords)
    distances_x, indices_x = nbrs_x.kneighbors(starred_coords)
    
    tier2_indices = set()
    for idx_array in indices_x:
        # Convert array positions to DataFrame indices
        tier2_indices.update([df_plot_index_list[i] for i in idx_array])
    # Remove tier 1 points (they're already in tier 1)
    tier2_indices = tier2_indices - tier1_indices
    
    # Tier 3: Find minPts_y nearest neighbors for each starred point
    nbrs_y = NearestNeighbors(n_neighbors=minPts_y, algorithm='auto').fit(all_coords)
    distances_y, indices_y = nbrs_y.kneighbors(starred_coords)
    
    tier3_indices = set()
    for idx_array in indices_y:
        # Convert array positions to DataFrame indices
        tier3_indices.update([df_plot_index_list[i] for i in idx_array])
    # Remove tier 1 and tier 2 points from tier 3
    tier3_indices = tier3_indices - tier1_indices - tier2_indices
    
    print(f"  Tier 1 (starred): {len(tier1_indices)} points")
    print(f"  Tier 2 (minPts_{minPts_x} neighbors): {len(tier2_indices)} points")
    print(f"  Tier 3 (minPts_{minPts_y} neighbors): {len(tier3_indices)} points")
    print(f"  Total cluster size: {len(tier1_indices) + len(tier2_indices) + len(tier3_indices)} points")
    
    return tier1_indices, tier2_indices, tier3_indices, nbrs_x, df_plot_index_list


# ==================== DATA EXPORT ====================

def export_cluster_data(df_subject, center_in_bin, radius_overlap_only, starred_points,
                       tier1_indices, tier2_indices, tier3_indices, 
                       minPts_x, minPts_y, subject, laterality, 
                       x_bin_index, y_bin_index, output_dir):
    """
    Export cluster membership data to CSV files.
    
    Parameters:
    -----------
    df_subject : DataFrame
        Original subject data
    center_in_bin, radius_overlap_only, starred_points : DataFrame
        Selected point sets
    tier1_indices, tier2_indices, tier3_indices : set
        Cluster tier memberships
    minPts_x, minPts_y : int
        Clustering parameters
    subject, laterality : str
        Subject identifiers
    x_bin_index, y_bin_index : int
        Bin indices
    output_dir : str
        Output directory path
    """
    if not output_dir:
        return
        
    # Start with the original df_subject data for this subject/laterality
    output_df = df_subject.copy()
    
    # Add cluster membership columns (initialize to False)
    output_df['is_center_in_bin'] = False
    output_df['is_radius_overlap'] = False
    output_df['is_starred'] = False
    output_df[f'in_cluster_minPts_{minPts_x}'] = False
    output_df[f'in_cluster_minPts_{minPts_y}'] = False
    output_df['cluster_tier'] = None
    
    # Mark different types of points
    output_df.loc[output_df.index.isin(center_in_bin.index), 'is_center_in_bin'] = True
    output_df.loc[output_df.index.isin(radius_overlap_only.index), 'is_radius_overlap'] = True
    output_df.loc[output_df.index.isin(starred_points.index), 'is_starred'] = True
    
    # Mark minPts_x cluster (tier 1 + tier 2)
    all_cluster_indices = tier1_indices | tier2_indices | tier3_indices
    minPts_x_cluster = tier1_indices | tier2_indices
    output_df.loc[output_df.index.isin(minPts_x_cluster), f'in_cluster_minPts_{minPts_x}'] = True
    
    # Mark minPts_y cluster (all three tiers)
    output_df.loc[output_df.index.isin(all_cluster_indices), f'in_cluster_minPts_{minPts_y}'] = True
    
    # Assign tier labels
    output_df.loc[output_df.index.isin(tier1_indices), 'cluster_tier'] = 'tier1_starred'
    output_df.loc[output_df.index.isin(tier2_indices), 'cluster_tier'] = f'tier2_minPts_{minPts_x}'
    output_df.loc[output_df.index.isin(tier3_indices), 'cluster_tier'] = f'tier3_minPts_{minPts_y}'
    
    # Save full dataframe to CSV
    csv_filename = os.path.join(
        output_dir,
        f"{subject}_{laterality}_binX{x_bin_index}_binY{y_bin_index}_CLUSTER_DATA.csv"
    )
    output_df.to_csv(csv_filename, index=True)
    print(f"\n  Saved full data: {csv_filename}")
    
    # Create filtered dataframe with only cluster members
    cluster_filter = (
        output_df['is_starred'] | 
        output_df[f'in_cluster_minPts_{minPts_x}'] | 
        output_df[f'in_cluster_minPts_{minPts_y}']
    )
    output_df_filtered = output_df[cluster_filter].copy()
    
    csv_filename_filtered = os.path.join(
        output_dir,
        f"{subject}_{laterality}_binX{x_bin_index}_binY{y_bin_index}_CLUSTER_DATA_FILTERED.csv"
    )
    output_df_filtered.to_csv(csv_filename_filtered, index=True)
    print(f"  Saved filtered data: {csv_filename_filtered}")


# ==================== 2D PLOTTING ====================

def create_2d_plot_with_radii(df_plot, starred_points, col_x, col_persistence,
                              x_bins, y_bins, x_bin_index, y_bin_index,
                              radius_size, show_all_radii, show_selected_radii_only,
                              discrete_rainbow, minPts_x, subject, laterality):
    """
    Create 2D persistence plot with radius circles.
    
    Returns:
    --------
    go.Figure : Plotly figure object
    """
    #print(f"\n  Creating 2D plot with radii...")
    
    fig_2d = go.Figure()

    # Determine which points to draw radii around
    if show_selected_radii_only and not starred_points.empty:
        points_with_radii = starred_points
    elif show_all_radii:
        points_with_radii = df_plot
    else:
        points_with_radii = pd.DataFrame()  # Empty

    # Add circles (radii) around points
    theta = np.linspace(0, 2*np.pi, 50)
    
    for idx, row in points_with_radii.iterrows():
        x_center = row[col_x]
        y_center = row[col_persistence]
        
        # Create circle coordinates
        x_circle = x_center + radius_size * np.cos(theta)
        y_circle = y_center + radius_size * np.sin(theta)
        
        # Determine color based on bin
        bin_idx = row['x_bin']
        if pd.notna(bin_idx):
            circle_color = discrete_rainbow[int(bin_idx)]
        else:
            circle_color = "gray"
        
        # Add filled circle
        fig_2d.add_trace(go.Scatter(
            x=x_circle,
            y=y_circle,
            mode='lines',
            fill='toself',
            fillcolor=circle_color,
            opacity=0.2,
            line=dict(color=circle_color, width=1),
            showlegend=False,
            hoverinfo='skip'
        ))

    # Add all points as scatter
    colors = [
        discrete_rainbow[int(b)] if not np.isnan(b) else "gray"
        for b in df_plot['x_bin']
    ]

    fig_2d.add_trace(go.Scatter(
        x=df_plot[col_x],
        y=df_plot[col_persistence],
        mode="markers",
        marker=dict(color=colors, opacity=0.7, size=6),
        name="All Points",
        hovertemplate=(
            "Radius: %{x:.4f}<br>"
            "Persistence: %{y:.4f}<br>"
            "<extra></extra>"
        )
    ))

    # Add starred points (center in bin OR radius overlap)
    if not starred_points.empty:
        fig_2d.add_trace(go.Scatter(
            x=starred_points[col_x],
            y=starred_points[col_persistence],
            mode="markers",
            marker=dict(color="red", size=12, symbol="star", 
                       line=dict(color='white', width=1)),
            name=f"Selected/Overlapping (n={len(starred_points)})",
            hovertemplate=(
                "<b>SELECTED/OVERLAPPING</b><br>"
                "Radius: %{x:.4f}<br>"
                "Persistence: %{y:.4f}<br>"
                "<extra></extra>"
            )
        ))

    # Add grid lines
    y_min, y_max = df_plot[col_persistence].min(), df_plot[col_persistence].max()
    x_min, x_max = df_plot[col_x].min(), df_plot[col_x].max()
    
    for i in range(len(x_bins)):
        fig_2d.add_shape(
            type="line",
            x0=x_bins[i], x1=x_bins[i],
            y0=y_min, y1=y_max,
            line=dict(color="black", width=1),
            opacity=0.3
        )
    for i in range(len(y_bins)):
        fig_2d.add_shape(
            type="line",
            x0=x_min, x1=x_max,
            y0=y_bins[i], y1=y_bins[i],
            line=dict(color="black", width=1),
            opacity=0.3
        )

    # Highlight selected bin
    bin_x_min = x_bins[x_bin_index]
    bin_x_max = x_bins[x_bin_index + 1]
    bin_y_min = y_bins[y_bin_index]
    bin_y_max = y_bins[y_bin_index + 1]
    
    fig_2d.add_shape(
        type="rect",
        x0=bin_x_min,
        x1=bin_x_max,
        y0=bin_y_min,
        y1=bin_y_max,
        line=dict(color="red", width=3),
        fillcolor="red",
        opacity=0.1
    )

    # Update layout
    title_suffix = f" [r={radius_size}"
    if show_selected_radii_only:
        title_suffix += ", radii on selected only]"
    elif show_all_radii:
        title_suffix += ", radii on all]"
    else:
        title_suffix += ", no radii]"
    
    fig_2d.update_layout(
        title=f"{subject} ({laterality}) | Bin X={x_bin_index}, Y={y_bin_index}{title_suffix}",
        xaxis_title=f"Radius @ minPts {minPts_x}",
        yaxis_title="Persistence (Δ Radius)",
        width=1000,
        height=800,
        template="plotly_white",
        hovermode='closest'
    )
    
    return fig_2d


# ==================== 3D PLOTTING ====================

def apply_global_color_binning(df_plot, col_x, global_color_bins):
    """Apply global color binning to the data - NOW USES 'i' VALUES."""
    df_plot['color_i'] = df_plot['i']  # Color by 'i' instead of radius
    df_plot['i_bin'] = pd.cut(
        df_plot['color_i'],
        bins=global_color_bins,
        labels=False,
        include_lowest=True,
        duplicates='drop'
    )
    return df_plot


def apply_local_color_binning(df_plot, col_x, local_color_bins):
    """Apply local color binning to the data - NOW USES 'i' VALUES."""
    df_plot['color_i'] = df_plot['i']  # Color by 'i' instead of radius
    df_plot['i_bin'] = pd.cut(
        df_plot['color_i'],
        bins=local_color_bins,
        labels=False,
        include_lowest=True,
        duplicates='drop'
    )
    return df_plot


def add_colorbar_trace(fig_3d, actual_n_bins, tick_labels, discrete_rainbow, scale_type="Global"):
    """
    Add invisible colorbar trace to 3D plot.
    
    Parameters:
    -----------
    scale_type : str
        Either "Global" or "Local" to indicate the colorbar type
    """
    fig_3d.add_trace(go.Scatter3d(
        x=[None],
        y=[None],
        z=[None],
        mode='markers',
        name='Colorbar',
        marker=dict(
            size=1,
            color=[0],
            colorscale=[
                [j / max(actual_n_bins - 1, 1), discrete_rainbow[j]]
                for j in range(actual_n_bins)
            ],
            cmin=0,
            cmax=max(actual_n_bins - 1, 0),
            showscale=True,
            colorbar=dict(
                title=f"Intensity ('i') ({scale_type})",  # Updated title
                tickvals=list(range(actual_n_bins)),
                ticktext=tick_labels,
                x=1.02,
                y=0.5,
                len=0.75
            ),
            opacity=0
        ),
        legendgroup='colorbar',
        showlegend=False,
        hoverinfo='skip'
    ))


def add_tier_legend(fig_3d, tier1_indices, tier2_indices, minPts_x):
    """Add tier shape legend to 3D plot."""
    tier_shapes = {
        'tier1': 'diamond',
        'tier2': 'circle',
    }
    
    tier_labels = {
        'tier1': f'◆ Starred (center/overlap, n={len(tier1_indices)})',
        'tier2': f'● minPts_{minPts_x} neighbors (n={len(tier2_indices)})',
    }
    
    for tier in ['tier1', 'tier2']:
        fig_3d.add_trace(go.Scatter3d(
            x=[None],
            y=[None],
            z=[None],
            mode='markers',
            name=tier_labels[tier],
            marker=dict(
                size=8,
                symbol=tier_shapes[tier],
                color='black',
                opacity=1
            ),
            legendgroup='tier_legend',
            showlegend=True,
            hoverinfo='skip'
        ))
    
    return tier_shapes


def add_cluster_connections(fig_3d, starred_points, nbrs_x, df_plot, df_plot_index_list):
    """Draw connections between points in each cluster (complete graph)."""
    #print(f"  Drawing connections for {len(starred_points)} starred point clusters...")
    
    for starred_idx, starred_row in starred_points.iterrows():
        starred_coord = starred_row[['x', 'y', 'z', 'i']].values.reshape(1, -1)
        
        # Find the minPts_x nearest neighbors for this specific starred point
        distances, indices = nbrs_x.kneighbors(starred_coord)
        
        # Get the DataFrame indices of the neighbors (including the starred point itself)
        neighbor_df_indices = [df_plot_index_list[i] for i in indices[0]]
        
        # Connect all points in this cluster to each other (complete graph)
        for i, idx1 in enumerate(neighbor_df_indices):
            for idx2 in neighbor_df_indices[i+1:]:  # Only connect each pair once
                row1 = df_plot.loc[idx1]
                row2 = df_plot.loc[idx2]
                
                fig_3d.add_trace(go.Scatter3d(
                    x=[row1['x'], row2['x']],
                    y=[row1['y'], row2['y']],
                    z=[row1['z'], row2['z']],
                    mode='lines',
                    line=dict(color='black', width=5),
                    showlegend=False,
                    legendgroup='cluster',
                    hoverinfo='skip',
                    opacity=0.5
                ))


def add_cluster_points(fig_3d, cluster_points, tier1_indices, tier2_indices,
                      tier_shapes, actual_n_bins_global, discrete_rainbow):
    """Add cluster point markers to 3D plot - NOW COLORED BY 'i' VALUE."""
    # Filter to only tier1 and tier2
    cluster_points_filtered = cluster_points[cluster_points['tier'].isin(['tier1', 'tier2'])].copy()
    all_subregions_in_cluster = sorted(cluster_points_filtered['subregion'].unique())
    
    for subregion in all_subregions_in_cluster:
        subregion_data = cluster_points_filtered[cluster_points_filtered['subregion'] == subregion]
        
        for tier in ['tier1', 'tier2']:
            tier_data = subregion_data[subregion_data['tier'] == tier]
            
            if tier_data.empty:
                continue

            fig_3d.add_trace(go.Scatter3d(
                x=tier_data['x'],
                y=tier_data['y'],
                z=tier_data['z'],
                mode='markers',
                name=f'{subregion} (cluster)',
                marker=dict(
                    size=8,
                    symbol=tier_shapes[tier],
                    color=tier_data['i_bin'],  # Changed from 'radius_bin' to 'i_bin'
                    colorscale=[
                        [j / max(actual_n_bins_global - 1, 1), discrete_rainbow[j]]
                        for j in range(actual_n_bins_global)
                    ],
                    cmin=0,
                    cmax=max(actual_n_bins_global - 1, 0),
                    showscale=False,
                    opacity=0.8
                ),
                legendgroup='cluster',
                showlegend=(tier == 'tier1'),
                hovertemplate=f"<b>{subregion}</b> (cluster)<br>" +
                              f"Tier: {tier}<br>" +
                              f"i value: %{{customdata[0]:.3f}}<br>" +  # Show actual 'i' value
                              f"i bin: %{{marker.color}}<br>" +
                              "x: %{x:.3f}<br>" +
                              "y: %{y:.3f}<br>" +
                              "z: %{z:.3f}<br>" +
                              "<extra></extra>",
                customdata=tier_data[['i']].values  # Pass 'i' values for hover
            ))


def add_background_points(fig_3d, df_plot, tier1_indices, tier2_indices,
                         actual_n_bins_global, discrete_rainbow):
    """Add background (non-cluster) points to 3D plot - NOW COLORED BY 'i' VALUE."""
    tier1_and_tier2_indices = tier1_indices | tier2_indices
    all_subregions_in_data = sorted(df_plot['subregion'].unique())
    
    for subregion in all_subregions_in_data:
        background_subregion = df_plot[
            (df_plot['subregion'] == subregion) & 
            (~df_plot.index.isin(tier1_and_tier2_indices))
        ]
        
        if not background_subregion.empty:
            fig_3d.add_trace(go.Scatter3d(
                x=background_subregion['x'],
                y=background_subregion['y'],
                z=background_subregion['z'],
                mode='markers',
                name=f'{subregion} (background)',
                marker=dict(
                    size=3,
                    color=background_subregion['i_bin'],  # Changed from 'radius_bin' to 'i_bin'
                    colorscale=[
                        [j / max(actual_n_bins_global - 1, 1), discrete_rainbow[j]]
                        for j in range(actual_n_bins_global)
                    ],
                    cmin=0,
                    cmax=max(actual_n_bins_global - 1, 0),
                    showscale=False,
                    opacity=0.3
                ),
                legendgroup=f'background_{subregion}',
                showlegend=True,
                hovertemplate=f"<b>{subregion}</b> (background)<br>" +
                              f"i value: %{{customdata[0]:.3f}}<br>" +  # Show actual 'i' value
                              f"i bin: %{{marker.color}}<br>" +
                              "x: %{x:.3f}<br>" +
                              "y: %{y:.3f}<br>" +
                              "z: %{z:.3f}<br>" +
                              "<extra></extra>",
                customdata=background_subregion[['i']].values  # Pass 'i' values for hover
            ))


def create_3d_cluster_plot(df_plot, cluster_points, starred_points,
                          tier1_indices, tier2_indices, tier3_indices,
                          nbrs_x, df_plot_index_list,
                          actual_n_bins, tick_labels, color_bins,
                          discrete_rainbow, minPts_x, subject, laterality,
                          x_bin_index, y_bin_index, scale_type="Global"):
    """
    Create 3D cluster visualization plot.
    
    Parameters:
    -----------
    scale_type : str
        Either "Global" or "Local" to indicate colorbar scaling
    
    Returns:
    --------
    go.Figure : Plotly figure object
    """
    #print(f"\n  Creating 3D cluster plot ({scale_type} scale)...")
    
    fig_3d = go.Figure()
    
    # Add colorbar
    add_colorbar_trace(fig_3d, actual_n_bins, tick_labels, discrete_rainbow, scale_type)
    
    # Add tier legend
    tier_shapes = add_tier_legend(fig_3d, tier1_indices, tier2_indices, minPts_x)
    
    # Add connections between cluster points
    add_cluster_connections(fig_3d, starred_points, nbrs_x, df_plot, df_plot_index_list)
    
    # Add cluster point markers
    add_cluster_points(fig_3d, cluster_points, tier1_indices, tier2_indices,
                      tier_shapes, actual_n_bins, discrete_rainbow)
    
    # Add background points
    add_background_points(fig_3d, df_plot, tier1_indices, tier2_indices,
                         actual_n_bins, discrete_rainbow)
    
    # Update layout
    fig_3d.update_layout(
        title=f"3D Cluster ({scale_type}): {subject} ({laterality}) | Bin X={x_bin_index}, Y={y_bin_index}",
        width=1100,
        height=800,
        scene=dict(
            xaxis_title='X',
            yaxis_title='Y',
            zaxis_title='Z'
        ),
        legend=dict(
            x=-0.15,
            y=0.5,
            xanchor="right",
            yanchor="middle",
            tracegroupgap=10
        )
    )
    
    return fig_3d


# ==================== MAIN PROCESSING FUNCTION ====================

def plot_2d_with_radii_and_3d_clusters(
    wide_df,
    minPts_x,
    minPts_y,
    x_bin_index,
    y_bin_index,
    output_dir=None,
    n_bins=8,
    n_color_bins=8,
    radius_size=0.001,
    show_all_radii=True,
    show_selected_radii_only=False
):
    """
    Create 2D persistence plots with radius overlap detection and 3D clustering plots.
    
    This function:
    1. Identifies points whose center is in the selected bin
    2. Identifies additional points whose radius overlaps the selected bin
    3. Creates 3D plots with three-tier nearest neighbor clustering
    4. Exports cluster membership data to CSV
    
    NOTE: Points in 3D plots are now colored by their 'i' value instead of radius.
    
    Parameters:
    -----------
    wide_df : DataFrame
        DataFrame with pivoted radius data
    minPts_x : int
        minPts value for x-axis and tier 2 clustering
    minPts_y : int
        minPts value for y-axis and tier 3 clustering
    x_bin_index : int
        X bin to highlight
    y_bin_index : int
        Y bin to highlight
    output_dir : str, optional
        Directory to save HTML and CSV outputs
    n_bins : int
        Number of bins for x and y axes (default 8)
    n_color_bins : int
        Number of color bins for 3D visualization (default 8)
    radius_size : float
        Radius to draw around each point (default 0.001)
    show_all_radii : bool
        If True, show radii around all points in 2D plot (default True)
    show_selected_radii_only : bool
        If True, only show radii around selected/overlapping points (default False)
    """
    
    if output_dir:
        # Create a subfolder based on minPts and bin values
        subfolder_name = f"minPts_{minPts_x}_{minPts_y}_binX{x_bin_index}_binY{y_bin_index}"
        output_dir = os.path.join(output_dir, subfolder_name)
        os.makedirs(output_dir, exist_ok=True)
        print(f"\nOutput directory: {output_dir}")

    col_x = f"radius_minPts_{minPts_x}"
    col_y_raw = f"radius_minPts_{minPts_y}"
    col_persistence = "persistence_delta"

    # Check required columns
    required_cols = [
        col_x, col_y_raw, 'x', 'y', 'z', 'i',
        'Subject', 'Laterality', 'subregion'
    ]
    missing = [c for c in required_cols if c not in wide_df.columns]
    if missing:
        raise ValueError(f"Missing columns: {missing}")

    # Calculate persistence
    wide_df = wide_df.copy()
    wide_df[col_persistence] = wide_df[col_y_raw] - wide_df[col_x]

    # Get color palette
    discrete_rainbow = get_color_palette()

    # Calculate global colorbar parameters (now based on 'i' values)
    (global_i_min, global_i_max, actual_n_bins_global, 
     global_color_bins, global_tick_labels) = calculate_global_colorbar_params(
        wide_df, col_x, n_color_bins
    )

    # Loop through subjects and lateralities
    for (subject, laterality), df_subject in wide_df.groupby(['Subject', 'Laterality']):

        print(f"\n{'='*60}")
        print(f"Processing {subject} ({laterality})...")
        print(f"{'='*60}")

        df_subject = df_subject.copy()
        df_plot = df_subject[
            [col_x, col_persistence, 'x', 'y', 'z', 'i', 'subregion']
        ].dropna().copy()

        if df_plot.empty:
            print(f"  No valid data for {subject} ({laterality}), skipping...")
            continue

        # Create 2D bins
        x_bins, y_bins = create_2d_bins(df_plot, col_x, col_persistence, n_bins)
        
        # Assign bins to data
        df_plot['x_bin'] = pd.cut(
            df_plot[col_x],
            bins=x_bins,
            labels=False,
            include_lowest=True
        )
        df_plot['y_bin'] = pd.cut(
            df_plot[col_persistence],
            bins=y_bins,
            labels=False,
            include_lowest=True
        )

        # Select starred points (center in bin or radius overlap)
        center_in_bin, radius_overlap_only, starred_points = select_starred_points(
            df_plot, col_x, col_persistence, x_bins, y_bins,
            x_bin_index, y_bin_index, radius_size
        )
        
        if starred_points.empty:
            print(f"  Warning: No points found in or overlapping bin X={x_bin_index}, Y={y_bin_index}")
            print(f"  Skipping this subject/laterality...")
            continue

        # Compute nearest neighbor clustering
        tier1_indices, tier2_indices, tier3_indices, nbrs_x, df_plot_index_list = \
            compute_nearest_neighbor_tiers(df_plot, starred_points, minPts_x, minPts_y)
        
        # Collect all cluster points
        all_cluster_indices = tier1_indices | tier2_indices | tier3_indices
        cluster_points = df_plot.loc[list(all_cluster_indices)].copy()
        
        # Assign tier labels
        cluster_points['tier'] = 'tier3'  # default
        cluster_points.loc[cluster_points.index.isin(tier2_indices), 'tier'] = 'tier2'
        cluster_points.loc[cluster_points.index.isin(tier1_indices), 'tier'] = 'tier1'

        # Export cluster data to CSV
        export_cluster_data(
            df_subject, center_in_bin, radius_overlap_only, starred_points,
            tier1_indices, tier2_indices, tier3_indices,
            minPts_x, minPts_y, subject, laterality,
            x_bin_index, y_bin_index, output_dir
        )

        # Create 2D plot with radii
        fig_2d = create_2d_plot_with_radii(
            df_plot, starred_points, col_x, col_persistence,
            x_bins, y_bins, x_bin_index, y_bin_index,
            radius_size, show_all_radii, show_selected_radii_only,
            discrete_rainbow, minPts_x, subject, laterality
        )

        # Save 2D plot
        if output_dir:
            filename_2d = f"{subject}_{laterality}_binX{x_bin_index}_binY{y_bin_index}_2D_WITH_RADII.html"
            filepath_2d = os.path.join(output_dir, filename_2d)
            fig_2d.write_html(filepath_2d)
            #print(f"  Saved 2D plot: {filename_2d}")

        # ==================== CREATE 3D PLOTS (GLOBAL AND LOCAL) ====================
        
        # Calculate local colorbar parameters for this subject (now based on 'i' values)
        (local_i_min, local_i_max, actual_n_bins_local,
         local_color_bins, local_tick_labels) = calculate_local_colorbar_params(
            df_plot, col_x, n_color_bins
        )
        #print(f"  Local 'i' range: {local_i_min:.4f} to {local_i_max:.4f}")
        
        # --- GLOBAL SCALE 3D PLOT ---
        # Apply global color binning (now based on 'i' values)
        df_plot_global = apply_global_color_binning(df_plot.copy(), col_x, global_color_bins)
        
        # Update cluster_points with global color binning
        cluster_points_global = df_plot_global.loc[list(all_cluster_indices)].copy()
        cluster_points_global['tier'] = 'tier3'
        cluster_points_global.loc[cluster_points_global.index.isin(tier2_indices), 'tier'] = 'tier2'
        cluster_points_global.loc[cluster_points_global.index.isin(tier1_indices), 'tier'] = 'tier1'

        # Create 3D cluster plot with global scale
        fig_3d_global = create_3d_cluster_plot(
            df_plot_global, cluster_points_global, starred_points,
            tier1_indices, tier2_indices, tier3_indices,
            nbrs_x, df_plot_index_list,
            actual_n_bins_global, global_tick_labels, global_color_bins,
            discrete_rainbow, minPts_x, subject, laterality,
            x_bin_index, y_bin_index, scale_type="Global"
        )

        # Save global 3D plot
        if output_dir:
            filename_3d_global = f"{subject}_{laterality}_binX{x_bin_index}_binY{y_bin_index}_3D_CLUSTER_GLOBAL.html"
            filepath_3d_global = os.path.join(output_dir, filename_3d_global)
            fig_3d_global.write_html(filepath_3d_global)
            #print(f"  Saved 3D plot (Global): {filename_3d_global}")
        
        # --- LOCAL SCALE 3D PLOT ---
        # Apply local color binning (now based on 'i' values)
        df_plot_local = apply_local_color_binning(df_plot.copy(), col_x, local_color_bins)
        
        # Update cluster_points with local color binning
        cluster_points_local = df_plot_local.loc[list(all_cluster_indices)].copy()
        cluster_points_local['tier'] = 'tier3'
        cluster_points_local.loc[cluster_points_local.index.isin(tier2_indices), 'tier'] = 'tier2'
        cluster_points_local.loc[cluster_points_local.index.isin(tier1_indices), 'tier'] = 'tier1'

        # Create 3D cluster plot with local scale
        fig_3d_local = create_3d_cluster_plot(
            df_plot_local, cluster_points_local, starred_points,
            tier1_indices, tier2_indices, tier3_indices,
            nbrs_x, df_plot_index_list,
            actual_n_bins_local, local_tick_labels, local_color_bins,
            discrete_rainbow, minPts_x, subject, laterality,
            x_bin_index, y_bin_index, scale_type="Local"
        )

        # Save local 3D plot
        if output_dir:
            filename_3d_local = f"{subject}_{laterality}_binX{x_bin_index}_binY{y_bin_index}_3D_CLUSTER_LOCAL.html"
            filepath_3d_local = os.path.join(output_dir, filename_3d_local)
            fig_3d_local.write_html(filepath_3d_local)
            #print(f"  Saved 3D plot (Local): {filename_3d_local}")

    print("\n" + "="*60)
    print("✓ Completed processing all subjects!")
    print("="*60)


# ==================== EXAMPLE USAGE ====================

if __name__ == "__main__":
    # Load your data
    dbscan_data = pd.read_csv("/Users/ajayshenoy/Downloads/AShenoy_data_directory/Research/Radii_Python_newData/newData_combined_coreRadii_all_minPts.csv")
    
    # List of columns that define each unique point
    index_cols = [
        'x', 'y', 'z', 'i', 'subregion', 'id', 'Image', 'Sex',
        'Genotype', 'Brain', 'Laterality', 'Region', 'Subject'
    ]
    
    # Pivot: minPts becomes columns, recorded_radius as values
    wide_df = dbscan_data.pivot_table(
        index=index_cols,
        columns='minPts',
        values='recorded_radius'
    )
    
    # Rename columns (3 → radius_minPts_3, etc.)
    wide_df.columns = [f"radius_minPts_{col}" for col in wide_df.columns]
    
    # Convert pivot index back into columns
    wide_df = wide_df.reset_index()
    
    # Create plots with radii and 3D clusters
    plot_2d_with_radii_and_3d_clusters(
        wide_df=wide_df,
        minPts_x=3,
        minPts_y=5,
        x_bin_index=3,
        y_bin_index=1,
        output_dir="/Users/ajayshenoy/Downloads/AShenoy_data_directory/Research/Radii_Python_newData/filtered_images/intensity",
        n_bins=8,
        n_color_bins=8,
        radius_size=0.001,  # Radius around each point in 2D
        show_all_radii=True,  # Show radii around all points
        show_selected_radii_only=False  # Set to True to only show radii on selected/overlapping points
    )
```

```python
import pandas as pd
import numpy as np
import os
from itertools import product
from sklearn.neighbors import NearestNeighbors


def circle_overlaps_rectangle(x_center, y_center, radius, rect_x_min, rect_x_max, rect_y_min, rect_y_max):
    """Check if a circle overlaps with a rectangle."""
    closest_x = np.clip(x_center, rect_x_min, rect_x_max)
    closest_y = np.clip(y_center, rect_y_min, rect_y_max)
    distance = np.sqrt((x_center - closest_x)**2 + (y_center - closest_y)**2)
    return distance <= radius


def create_2d_bins(df_plot, col_x, col_persistence, n_bins):
    """Create bins for 2D persistence plot."""
    x_min, x_max = df_plot[col_x].min(), df_plot[col_x].max()
    y_min, y_max = df_plot[col_persistence].min(), df_plot[col_persistence].max()
    x_bins = np.linspace(x_min, x_max, n_bins + 1)
    y_bins = np.linspace(y_min, y_max, n_bins + 1)
    return x_bins, y_bins


def select_starred_points(df_plot, col_x, col_persistence, x_bins, y_bins,
                          x_bin_index, y_bin_index, radius_size):
    """Select points whose center is in the bin or whose radius overlaps the bin."""
    bin_x_min = x_bins[x_bin_index]
    bin_x_max = x_bins[x_bin_index + 1]
    bin_y_min = y_bins[y_bin_index]
    bin_y_max = y_bins[y_bin_index + 1]

    df_plot['radius_overlaps_bin'] = df_plot.apply(
        lambda row: circle_overlaps_rectangle(
            row[col_x], row[col_persistence], radius_size,
            bin_x_min, bin_x_max, bin_y_min, bin_y_max
        ),
        axis=1
    )

    starred_points = df_plot[
        ((df_plot['x_bin'] == x_bin_index) & (df_plot['y_bin'] == y_bin_index)) |
        df_plot['radius_overlaps_bin']
    ].copy()

    return starred_points


def compute_cluster_at_minPts(df_plot, starred_points, minPts):
    """
    Compute cluster membership for a given minPts value.
    Returns set of indices that are in the cluster.
    """
    if starred_points.empty:
        return set()

    all_coords = df_plot[['x', 'y', 'z', 'i']].values
    starred_coords = starred_points[['x', 'y', 'z', 'i']].values
    df_plot_index_list = df_plot.index.tolist()

    tier1_indices = set(starred_points.index)

    nbrs = NearestNeighbors(n_neighbors=minPts, algorithm='auto').fit(all_coords)
    distances, indices = nbrs.kneighbors(starred_coords)

    tier2_indices = set()
    for idx_array in indices:
        tier2_indices.update([df_plot_index_list[i] for i in idx_array])

    cluster_indices = tier1_indices | tier2_indices
    return cluster_indices


def analyze_cluster_subregions(df_plot, cluster_indices):
    """
    Analyze subregion composition of a cluster.
    """
    if not cluster_indices:
        return {
            'total_points': 0,
            'subregion_counts': {},
            'subregion_percentages': {},
            'primary_subregion': None,
            'primary_percentage': 0.0,
            'is_pure': False,
            'num_subregions': 0
        }

    cluster_points = df_plot.loc[list(cluster_indices)]
    total_points = len(cluster_points)
    subregion_counts = cluster_points['subregion'].value_counts().to_dict()
    subregion_percentages = {
        subregion: (count / total_points * 100)
        for subregion, count in subregion_counts.items()
    }
    primary_subregion = max(subregion_counts.items(), key=lambda x: x[1])[0]
    primary_percentage = subregion_percentages[primary_subregion]
    is_pure = len(subregion_counts) == 1

    return {
        'total_points': total_points,
        'subregion_counts': subregion_counts,
        'subregion_percentages': subregion_percentages,
        'primary_subregion': primary_subregion,
        'primary_percentage': primary_percentage,
        'is_pure': is_pure,
        'num_subregions': len(subregion_counts)
    }


def analyze_minPts_pair_batch(
    wide_df,
    minPts_pairs,
    n_bins=8,
    radius_size=0.001
):
    """
    Analyze a batch of minPts pairs.
    This is the core processing function that handles one batch.
    """
    results = []

    for minPts_x, minPts_y in minPts_pairs:
        print(f"\n{'='*80}")
        print(f"Processing minPts pair: ({minPts_x}, {minPts_y})")
        print(f"{'='*80}")

        col_x = f"radius_minPts_{minPts_x}"
        col_y_raw = f"radius_minPts_{minPts_y}"
        col_persistence = "persistence_delta"

        if col_x not in wide_df.columns or col_y_raw not in wide_df.columns:
            print(f"  Warning: Columns {col_x} or {col_y_raw} not found, skipping...")
            continue

        # Calculate persistence — scoped per pair, outside subject loop
        df_temp = wide_df.copy()
        df_temp[col_persistence] = df_temp[col_y_raw] - df_temp[col_x]

        for (subject, laterality), df_subject in df_temp.groupby(['Subject', 'Laterality']):

            print(f"\n  Subject: {subject} ({laterality})")

            df_plot = df_subject[
                [col_x, col_persistence, 'x', 'y', 'z', 'i', 'subregion']
            ].dropna().copy()

            if df_plot.empty:
                print(f"    No valid data, skipping...")
                continue

            x_bins, y_bins = create_2d_bins(df_plot, col_x, col_persistence, n_bins)

            x_range = df_plot[col_x].max() - df_plot[col_x].min()
            y_range = df_plot[col_persistence].max() - df_plot[col_persistence].min()

            if x_range == 0 or y_range == 0:
                print(f"    Warning: Insufficient variance (x_range={x_range:.6f}, y_range={y_range:.6f}), skipping...")
                continue

            try:
                df_plot['x_bin'] = pd.cut(
                    df_plot[col_x],
                    bins=x_bins,
                    labels=False,
                    include_lowest=True,
                    duplicates='drop'
                )
                df_plot['y_bin'] = pd.cut(
                    df_plot[col_persistence],
                    bins=y_bins,
                    labels=False,
                    include_lowest=True,
                    duplicates='drop'
                )
            except Exception as e:
                print(f"    Warning: Error creating bins: {e}, skipping...")
                continue

            if df_plot['x_bin'].isna().all() or df_plot['y_bin'].isna().all():
                print(f"    Warning: All bins are NA, skipping...")
                continue

            for x_bin_idx in range(int(df_plot['x_bin'].max()) + 1):
                for y_bin_idx in range(int(df_plot['y_bin'].max()) + 1):

                    starred_points = select_starred_points(
                        df_plot, col_x, col_persistence, x_bins, y_bins,
                        x_bin_idx, y_bin_idx, radius_size
                    )

                    if starred_points.empty:
                        results.append({
                            'subject': subject,
                            'laterality': laterality,
                            'minPts_x': minPts_x,
                            'minPts_y': minPts_y,
                            'x_bin': x_bin_idx,
                            'y_bin': y_bin_idx,
                            'num_starred_points': 0,
                            'total_cluster_points_x': 0,
                            'num_subregions_x': 0,
                            'primary_subregion_x': None,
                            'primary_percentage_x': 0.0,
                            'is_pure_x': False,
                            'total_cluster_points_y': 0,
                            'num_subregions_y': 0,
                            'primary_subregion_y': None,
                            'primary_percentage_y': 0.0,
                            'is_pure_y': False,
                        })
                        continue

                    # Compute cluster at BOTH minPts values
                    cluster_indices_x = compute_cluster_at_minPts(df_plot, starred_points, minPts_x)
                    cluster_indices_y = compute_cluster_at_minPts(df_plot, starred_points, minPts_y)

                    analysis_x = analyze_cluster_subregions(df_plot, cluster_indices_x)
                    analysis_y = analyze_cluster_subregions(df_plot, cluster_indices_y)

                    result_record = {
                        'subject': subject,
                        'laterality': laterality,
                        'minPts_x': minPts_x,
                        'minPts_y': minPts_y,
                        'x_bin': x_bin_idx,
                        'y_bin': y_bin_idx,
                        'num_starred_points': len(starred_points),
                        # Results for minPts_x clustering
                        'total_cluster_points_x': analysis_x['total_points'],
                        'num_subregions_x': analysis_x['num_subregions'],
                        'primary_subregion_x': analysis_x['primary_subregion'],
                        'primary_percentage_x': analysis_x['primary_percentage'],
                        'is_pure_x': analysis_x['is_pure'],
                        # Results for minPts_y clustering
                        'total_cluster_points_y': analysis_y['total_points'],
                        'num_subregions_y': analysis_y['num_subregions'],
                        'primary_subregion_y': analysis_y['primary_subregion'],
                        'primary_percentage_y': analysis_y['primary_percentage'],
                        'is_pure_y': analysis_y['is_pure'],
                    }

                    for subregion, count in analysis_x['subregion_counts'].items():
                        result_record[f'count_x_{subregion}'] = count
                        result_record[f'pct_x_{subregion}'] = analysis_x['subregion_percentages'][subregion]

                    for subregion, count in analysis_y['subregion_counts'].items():
                        result_record[f'count_y_{subregion}'] = count
                        result_record[f'pct_y_{subregion}'] = analysis_y['subregion_percentages'][subregion]

                    results.append(result_record)

        # BUG FIX 2: del df_temp must be outside the subject loop, inside the pair loop
        del df_temp

    # BUG FIX 3: return must be outside both loops
    return pd.DataFrame(results)


def analyze_all_combinations_batched(
    wide_df,
    minPts_pairs,
    n_bins=8,
    radius_size=0.001,
    output_dir=None,
    batch_size=10
):
    """
    Analyze cluster composition across all combinations of minPts pairs and bins.
    Processes in batches and saves incrementally to avoid memory issues.

    This version STOPS after creating all batch CSV files.
    To combine batches later, run: combine_batch_results(output_dir)
    """
    if output_dir:
        os.makedirs(output_dir, exist_ok=True)

    total_pairs = len(minPts_pairs)
    num_batches = (total_pairs + batch_size - 1) // batch_size

    print(f"\n{'='*80}")
    print(f"BATCH PROCESSING SETUP")
    print(f"{'='*80}")
    print(f"Total minPts pairs: {total_pairs}")
    print(f"Batch size: {batch_size}")
    print(f"Number of batches: {num_batches}")
    print(f"{'='*80}\n")

    for batch_num in range(num_batches):
        start_idx = batch_num * batch_size
        end_idx = min((batch_num + 1) * batch_size, total_pairs)
        batch_pairs = minPts_pairs[start_idx:end_idx]

        print(f"\n{'#'*80}")
        print(f"BATCH {batch_num + 1}/{num_batches}")
        print(f"Processing pairs {start_idx + 1} to {end_idx} of {total_pairs}")
        print(f"{'#'*80}")

        batch_results = analyze_minPts_pair_batch(
            wide_df=wide_df,
            minPts_pairs=batch_pairs,
            n_bins=n_bins,
            radius_size=radius_size
        )

        if output_dir and not batch_results.empty:
            batch_filename = f'batch_{batch_num + 1:03d}_results.csv'
            batch_path = os.path.join(output_dir, batch_filename)
            batch_results.to_csv(batch_path, index=False)
            print(f"\n{'='*80}")
            print(f"Saved batch {batch_num + 1} results to: {batch_filename}")
            print(f"Records in batch: {len(batch_results)}")
            print(f"{'='*80}")

        del batch_results

    print(f"\n{'#'*80}")
    print(f"ALL BATCHES COMPLETE!")
    print(f"Batch CSV files saved to: {output_dir}")
    print(f"To combine batches, run: combine_batch_results('{output_dir}')")
    print(f"{'#'*80}")


def combine_batch_results(output_dir):
    """
    Combine all batch result files into a single consolidated file.
    """
    import glob

    batch_files = sorted(glob.glob(os.path.join(output_dir, 'batch_*_results.csv')))

    if not batch_files:
        print("No batch files found to combine.")
        return

    print(f"Found {len(batch_files)} batch files")

    all_results = []
    for batch_file in batch_files:
        df = pd.read_csv(batch_file)
        all_results.append(df)
        print(f"  Loaded: {os.path.basename(batch_file)} ({len(df)} records)")

    combined_df = pd.concat(all_results, ignore_index=True)

    combined_path = os.path.join(output_dir, 'cluster_subregion_analysis_full.csv')
    combined_df.to_csv(combined_path, index=False)

    print(f"\n{'='*80}")
    print(f"Combined results saved to: cluster_subregion_analysis_full.csv")
    print(f"Total records: {len(combined_df)}")
    print(f"{'='*80}")

    create_summary_statistics(combined_df, output_dir)

    return combined_df


def create_summary_statistics(results_df, output_dir):
    """Create and save summary statistics from the analysis results.
    Updated to handle _x and _y suffixed columns.
    """
    # BUG FIX 4: filter on _x variant (both _x and _y will be 0 together for empty bins)
    clusters_df = results_df[results_df['total_cluster_points_x'] > 0].copy()

    if clusters_df.empty:
        print("No clusters found in analysis.")
        return

    # Summary 1: By minPts pair — aggregate both _x and _y columns
    print("\n" + "="*80)
    print("SUMMARY BY minPts PAIR")
    print("="*80)

    minPts_summary = clusters_df.groupby(['minPts_x', 'minPts_y']).agg({
        'total_cluster_points_x': ['count', 'mean', 'std', 'min', 'max'],
        'total_cluster_points_y': ['mean', 'std', 'min', 'max'],
        'num_subregions_x': ['mean', 'std'],
        'num_subregions_y': ['mean', 'std'],
        'is_pure_x': 'mean',
        'is_pure_y': 'mean',
        'primary_percentage_x': ['mean', 'std'],
        'primary_percentage_y': ['mean', 'std'],
    }).round(2)

    print(minPts_summary)
    minPts_summary.to_csv(os.path.join(output_dir, 'summary_by_minPts.csv'))

    # Summary 2: By primary subregion (for both _x and _y)
    print("\n" + "="*80)
    print("SUMMARY BY PRIMARY SUBREGION (minPts_x clustering)")
    print("="*80)

    subregion_summary_x = clusters_df.groupby('primary_subregion_x').agg({
        'total_cluster_points_x': ['count', 'mean', 'std'],
        'is_pure_x': 'mean',
        'primary_percentage_x': ['mean', 'std']
    }).round(2)
    print(subregion_summary_x)
    subregion_summary_x.to_csv(os.path.join(output_dir, 'summary_by_subregion_x.csv'))

    print("\n" + "="*80)
    print("SUMMARY BY PRIMARY SUBREGION (minPts_y clustering)")
    print("="*80)

    subregion_summary_y = clusters_df.groupby('primary_subregion_y').agg({
        'total_cluster_points_y': ['count', 'mean', 'std'],
        'is_pure_y': 'mean',
        'primary_percentage_y': ['mean', 'std']
    }).round(2)
    print(subregion_summary_y)
    subregion_summary_y.to_csv(os.path.join(output_dir, 'summary_by_subregion_y.csv'))

    # Summary 3: By subject and laterality
    print("\n" + "="*80)
    print("SUMMARY BY SUBJECT")
    print("="*80)

    subject_summary = clusters_df.groupby(['subject', 'laterality']).agg({
        'total_cluster_points_x': ['count', 'mean'],
        'total_cluster_points_y': ['mean'],
        'num_subregions_x': 'mean',
        'num_subregions_y': 'mean',
        'is_pure_x': 'mean',
        'is_pure_y': 'mean',
    }).round(2)

    print(subject_summary)
    subject_summary.to_csv(os.path.join(output_dir, 'summary_by_subject.csv'))

    # Summary 4: Pure vs mixed clusters (for both)
    for suffix in ['x', 'y']:
        pure_col = f'is_pure_{suffix}'
        print(f"\n{'='*80}")
        print(f"PURE vs MIXED CLUSTERS (minPts_{suffix} clustering)")
        print("="*80)

        pure_counts = clusters_df[pure_col].value_counts()
        pure_pct = (pure_counts / len(clusters_df) * 100).round(2)

        print(f"Pure clusters: {pure_counts.get(True, 0)} ({pure_pct.get(True, 0)}%)")
        print(f"Mixed clusters: {pure_counts.get(False, 0)} ({pure_pct.get(False, 0)}%)")

        pure_summary = pd.DataFrame({
            'cluster_type': ['Pure', 'Mixed'],
            'count': [pure_counts.get(True, 0), pure_counts.get(False, 0)],
            'percentage': [pure_pct.get(True, 0), pure_pct.get(False, 0)]
        })
        pure_summary.to_csv(os.path.join(output_dir, f'summary_pure_vs_mixed_{suffix}.csv'), index=False)

    print(f"\n{'='*80}")
    print("Summary statistics saved!")
    print(f"{'='*80}")


# ==================== MAIN ====================

if __name__ == "__main__":

    print("Loading data...")
    dbscan_data = pd.read_csv("/Users/ajayshenoy/Downloads/AShenoy_data_directory/Research/Radii_Python_newData/newData_combined_coreRadii_all_minPts.csv")

    print("Pivoting data...")
    index_cols = [
        'x', 'y', 'z', 'i', 'subregion', 'id', 'Image', 'Sex',
        'Genotype', 'Brain', 'Laterality', 'Region', 'Subject'
    ]

    wide_df = dbscan_data.pivot_table(
        index=index_cols,
        columns='minPts',
        values='recorded_radius'
    )
    wide_df.columns = [f"radius_minPts_{col}" for col in wide_df.columns]
    wide_df = wide_df.reset_index()

    # Only pairs where minPts_y > minPts_x (no duplicates, no diagonal)
    minPts_values = list(range(3, 16))  # 3 through 15
    all_pairs = list(product(minPts_values, minPts_values))
    minPts_pairs = [(x, y) for x, y in all_pairs if y > x]

    print(f"Total minPts pairs to analyze: {len(minPts_pairs)}")
    print(f"Skipped pairs (where y <= x): {len(all_pairs) - len(minPts_pairs)}")
    print(f"MinPts range: {min(minPts_values)} to {max(minPts_values)}")

    print("\nStarting batched analysis...")
    analyze_all_combinations_batched(
        wide_df=wide_df,
        minPts_pairs=minPts_pairs,
        n_bins=8,
        radius_size=0.001,
        output_dir="/Users/ajayshenoy/Downloads/AShenoy_data_directory/Research/Radii_Python_newData/filtered_images/cluster_analysis_results",
        batch_size=10
    )

    print("\n" + "="*80)
    print("BATCH PROCESSING COMPLETE!")
    print("To combine batch files into one, run:")
    print("combine_batch_results('/path/to/output_dir')")
    print("="*80)
```
